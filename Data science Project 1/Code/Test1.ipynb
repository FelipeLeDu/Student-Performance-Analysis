### Exploration Analysis

primeiro teste: queremos saber se ter uma trabalho de meio perído afeta no desempenho de um aluno - Para isso temos duas variáveis, nota de cada aluno (numérica) e se tem um trabalho de meio período (em dummies). O que temos que fazer é separar cada nota por sua categoria e fazer um teste de hipótese comparando os grupos, sendo H0: as médias são iguais e HA: as médias são diferentes 

```python
from scipy.stats import shapiro
from scipy.stats import f
from scipy.stats import t

#First test: Does have a part-time job affect academic performance?

score_part_job = df_encoded[df_encoded["part_time_job_Yes"] == 1]["exam_score"]
score_no_part_job = df_encoded[df_encoded["part_time_job_Yes"] == 0]["exam_score"]

#Identify the distribution of this two variables 
plt.figure()
sns.histplot(score_part_job, kde=True, color="blue", label="Part-time job", stat="density")
plt.figure()
sns.histplot(score_no_part_job, kde=True, color="red", label="No-Part-time job", stat="density")

stat1, p1 = shapiro(score_no_part_job) #Shapiro-Wilk test for alpha = 0.05
stat2, p2 = shapiro(score_part_job) # h0: its normal distriburtion, ha: its not normal distribution
print(f"No Part time job: W={stat1:.3f}, p-value={p1:.3f}") # Se a variável mãe é normal, o que eu retirar dela vai ser normal também? 
print(f"Part time job: W={stat2:.3f}, p-value={p2:.3f}") # Is it possible to make a fuction to do this test for all the variables?

#The two variable are not normally distributes, but I will asume that they are btw. Also, we will 
#assume that the variables are idependent. 

#For the last, we need to now if the variables have equal variance.
# H0: v1 = v2 = v: theta = 0 # H1: v1 != v2, theta != 0 and alpha = 0.05
alpha = 0.05

n1 = len(score_no_part_job)
n2 = len(score_part_job)
df1 = n1 - 1
df2 = n2 - 1

S21 = score_no_part_job.var(ddof=1)
S22 = score_part_job.var(ddof=1)

# .ppf is inverse o cdf, so with .ppf we determine the critical value n% (cdf)
f2 = f.ppf(1-alpha/2, df1, df2) 
f1 = 1/f2
F_obs = S21/S22

if F_obs < f1 or F_obs > f2: 
    print("Reject H0: The variances are not equal")
else:
    print("Fail to reject H0: The variances are equal")

print(f"F_obs: {F_obs:.3f}, f1: {f1:.3f}, f2: {f2:.3f}")    
```

```python
#test to prove that having a part-time job affects academic performance with mean

Xmean = score_no_part_job.mean()
Ymean = score_part_job.mean()

Sxy = ((n1 - 1)*S21 + (n2 - 1)*S22)/(n1 + n2 - 2)
tc1 = t.ppf(alpha/2, n1 + n2 - 2)
tc2 = t.ppf(1 - alpha/2, n1 + n2 - 2)
T_obs = (Xmean - Ymean) / (np.sqrt(Sxy * (1/n1 + 1/n2)))
p_valor = 2 * t.cdf(-abs(T_obs), n1 + n2 - 2)

if T_obs < tc1 or T_obs > tc2: 
    print("Reject H0: Having a part-time job affects academic performance")
else: 
    print("Fail to reject H0: Having a part-time job does not affect academic performance, with alpha = 0.05")
print(f"T_obs: {T_obs:.3f}, tc1: {tc1:.3f}, tc2: {tc2:.3f}, p-value: {p_valor:.3f}")
```
